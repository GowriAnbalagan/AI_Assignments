```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int)
dataset=dataset.drop(columns=["sex_male","smoker_yes"])
dataset
```

```python
dataset.columns
```

```python
indep= dataset[['age', 'bmi', 'children', 'sex_female', 'smoker_no']]
indep
```

```python
dep=dataset[['charges']]
dep
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep, dep, test_size=0.30, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
#https://scikit-learn.org/1.5/modules/generated/sklearn.ensemble.RandomForestRegressor.html
#for hypertunning use some parameters
regressor=RandomForestRegressor(n_estimators=150,criterion='squared_error',random_state=0)
regressor=regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score= r2_score(y_test,y_pred) 
r_score
```

```python
import pickle
filename = "finalized_model_RF.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```
