```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int)
dataset=dataset.drop(columns=["sex_male","smoker_yes"])
dataset
```

```python
dataset.columns
```

```python
indep= dataset[['age', 'bmi', 'children', 'sex_female', 'smoker_no']]
indep
```

```python
dep=dataset[['charges']]
dep
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR

param_grid ={'criterion':['poisson','squared_error','friedman_mse','absolute_error'],'n_estimators':[10,100],'max_features':['sqrt','log2','int','auto']}

grid = GridSearchCV ( RandomForestRegressor(), param_grid, refit= True, verbose = 3, n_jobs = -1 )

#fitting the model for grid search
grid.fit(indep,dep)
```

```python
print( "The R_Score value for best parameter{}".format(grid.best_params_))
```

```python
re=grid.cv_results_
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_female_input=int(input("sex_female:"))
smoker_no_input=int(input("smoker_no:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_female_input,smoker_no_input]])
print("Future_Prediction={}".format(Future_Prediction))
```
